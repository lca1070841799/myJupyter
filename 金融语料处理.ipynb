```python
import pandas as pd
import numpy as np
```

```python
sheet=pd.read_excel('data/单轮语料-新网2020-04-03去重.xlsx')
```

```python
df=sheet.loc[:,['访客问题','机器人识别']]
df
```

```python
df=df.dropna(axis=0,how='any')
df
```

```python
#添加一列flag值
flag=np.ones(df.shape[0])
flag=flag.reshape(-1,1)
df_flag=pd.DataFrame(flag,columns=['flag'])
# print(flag.shape,df.shape,df_flag.shape)
# print(df_flag)
#合并df
df_res=df.join(df_flag)#默认按行索引进行合并

df_res['flag']=df_res['flag'].fillna(1)
df_res[['flag']]=df_res[['flag']].astype(int)
```
